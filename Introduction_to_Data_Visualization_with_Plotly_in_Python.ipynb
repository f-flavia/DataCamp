```python
# import data

pop_growth = pd.read_csv('pop_growth.csv', sep=';')
pop_growth.head(2)
```

```python
# to remove unnamed columns
pop_growth.drop(pop_growth.columns[pop_growth.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
pop_growth.head(2)
```

```python
# Create the basic figure
fig = go.Figure()

# Loop through the states
for state in ['NSW', 'QLD', 'VIC']:
  	# Subset the DataFrame
    df = pop_growth[pop_growth.State == state]
    # Add a trace for each state subset
    fig.add_trace(go.Bar(x=df['Local Government'], y=df['Change %'], name=state))

# Create the buttons
dropdown_buttons = [
{'label': "ALL", 'method': "update", 'args': [{"visible": [True, True, True]}, {"title": "ALL"}]},
{'label': "NSW", 'method': "update", 'args': [{"visible": [True, False, False]}, {"title": "NSW"}]},
{'label': "QLD", 'method': "update", 'args': [{"visible": [False, True, False]}, {"title": "QLD"}]},
{'label': "VIC", 'method': "update", 'args': [{"visible": [False, False, True]}, {"title": "VIC"}]},
]

# Update the figure to add dropdown menu
fig.update_layout({
  		'updatemenus': [
        {'active': 0, 'buttons': dropdown_buttons}
        ]})

# Show the plot
fig.show()
```

```python
# import data

house_prices = pd.read_csv('train.csv')
house_prices.head()
```

```python
house_prices['YearBuilt'].dtypes
```

```python
hp = house_prices.groupby(['YearBuilt','Neighborhood']).agg({'SalePrice':'mean'}).reset_index()
```

```python
hp_=pd.DataFrame(hp)
```

```python
hp_.head()
```

```python
# Create the basic figure
fig = go.Figure()

# Loop through the suburbs
for suburb in ['CollgCr', 'OldTown']:
  	# Subset the DataFrame
    df = hp_[hp_.Neighborhood == suburb]
    # Add a trace for each suburb subset
    fig.add_trace(go.Scatter(
                   x=df['YearBuilt'],
                   y=df['SalePrice'],
                   name=suburb, mode='lines'))
# Annotation
ga_annotation=[{ 'text': 'Price boom!','showarrow': True, 'x': 'Year: 1893', 'y': 325}]# NO DEU CERTO O Y

# Create the buttons
dropdown_buttons = [
{'label': "CollgCr", 'method': "update", 'args': [{"visible": [True, False]}, {'title': 'CollgCr', 'annotations': ga_annotation}]},
{'label': "OldTown", 'method': "update", 'args': [{"visible": [False, True]}, {'title': 'OldTown', 'annotations': []}]},
]

# Update the figure to add dropdown menu
fig.update_layout({
    'updatemenus':[{
            'type': 'dropdown',
            'x': 1.3, 'y': 0.5,
            'showactive': True, 'active': 0,
            'buttons': dropdown_buttons
            }]})

# Show the plot
fig.show()
```
